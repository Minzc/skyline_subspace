```python
f = open('C:\\scratch\\jiaxingl\\DBLPOnlyCitationOct19.txt', 'r')
s = f.read().split('\n\n')
author_index = {}
author_label = {}
c_index = {}
edge = {}
cnt = 0
c_cnt = 0

top_conf = set(['VLDB',
 'SIGMOD',
 'PODS',
 'ICDE',
 'ICDT',
 'EDBT',
 'KDD',
 'ICDM',
 'IJCAI',
 'AAAI',
 'ICML',
 'UAI',
 'UMAP',
 'NIPS',
 'AAMAS',
 'ACL',
 'AIED',
 'ITS',
 'SIGIR',
 'WWW',
 'ICIS',
 'PPoPP',
 'PACT',
 'IPDPS',
 'ICPP',
 'Euro-Par',
 'SIGGRAPH',
 'CVPR',
 'ICCV',
 'I3DG',
 'ACM-MM',
 'SIGCOMM',
 'PERFORMANCE',
 'SIGMETRICS',
 'INFOCOM',
 'MOBICOM',
 'IEE',
 'CCS',
 'SOSP',
 'OSDI',
 'FOCS',
 'STOC',
 'ICALP',
 'SODA',
 'ISMB'])

for i in top_conf:z
    c_index[i] = c_cnt
    c_cnt += 1

for paper in s:
    x = paper.split('\n')
    a_list = []
    for i in x:
        if len(i) < 2: continue
        if i[1] == '@':
            authors = i[2:].split(',')
            a_list = []
            for author in authors:
                if len(author) < 1: continue
                if author not in author_index:
                    author_index[author] = cnt
                    author_label[cnt] = {}
                    cnt += 1
                a_list.append(author_index[author])
            for x in a_list:
                for y in a_list:
                    if x < y:
                        edge[(x, y)] = 1
        if i[1] == 'c':
            if i[2:] not in c_index: continue
            label = c_index[i[2:]]
            for a_num in a_list:
                author_label[a_num][label] = 1

fa = open('C:\\scratch\\jiaxingl\\DBLP_author.txt', 'w')
for key, value in author_index.iteritems():
    fa.write(str(key) + ' ' + str(value) + ',\n')
fa.close()

fg = open('C:\\scratch\\jiaxingl\\DBLP_graph.txt', 'w')
for i, j in edge.keys():
    fg.write(str(i) + ' ' + str(j) + '\n')
fg.close()
fl = open('C:\\scratch\\jiaxingl\\DBLP_label.txt', 'w')
for key, value in author_label.iteritems():
    fl.write(str(len([str(i) for i in value.keys()])) + ' ' + ' '.join([str(i) for i in value.keys()])+'\n')
fl.close()
print cnt
print len(edge)
print c_cnt
```
